```python
import pandas as pd
product_df = pd.read_csv("product_link_0827.csv", encoding='cp949',index_col=0)
# 링크 CSV 파일 이름 확인 
```

```python
# 링크 혹은 다른 값이 비어있는 행 삭제
product_df = product_df.dropna().reset_index(drop=True)
```

```python
product_name = []
review_star = []
review_name = []
review_date = []
review_review = []
kkuk_lst = []


for i in range(len(product_df)):#len(product_df["link"])): 
    
    url = product_df["링크"][i] 
    
    driver = webdriver.Chrome("chromedriver.exe")
    try:
        driver.get(url)
        driver.implicitly_wait(3)

        iframe = driver.find_element(By.ID, "ifrmReview") 
        driver.switch_to.frame(iframe)

        # 평점 고르는 버튼 누르기
        try: # 리뷰가 없어서 버튼이 안 눌릴 경우에 대한 예외처리
            driver.find_element(By.XPATH, "/html/body/div[2]/ul/li[1]/div[1]/div[1]/button").send_keys(Keys.ENTER)
            time.sleep(random.uniform(0,2))

            ### 평점 1점 누르기
            driver.find_element(By.XPATH, "/html/body/div[2]/ul/li[1]/div[1]/div[2]/dl/dd[6]/label/span").click()
            time.sleep(random.uniform(0,2))

            p = 0
            while True:
                try:
                    driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/button").send_keys(Keys.ENTER)
                    # driver.implicitly_wait(3)
                    p = p + 1
                    # 리뷰 1000개당 15초 쉬기
                    if p % 50 == 0:
                        driver.execute_script("window.scrollBy(0,250);")
                        time.sleep(15)
                        print(p, "번째 스크롤 끝")
                # 더 누를 더보기 버튼이 없으면 몇번 눌렀는지 나오고 while문 빠져나오기
                except:
                    print(f"총 {p}번 더보기 버튼 누름")
                    break

            star = driver.find_elements(By.CSS_SELECTOR,'p.grade > span > em')
            name = driver.find_elements(By.CLASS_NAME, "name")
            date = driver.find_elements(By.CLASS_NAME, "date")
            review = driver.find_elements(By.CLASS_NAME, "cont_text_wrap")
            kkuk = driver.find_elements(By.CLASS_NAME,'kkukCount')



            # 리뷰 내용이 없을 수 있으므로
            # try except 문으로 처리하기
            for l in range(len(review)):
                try:
                    product_name.append(product_df["상품명"][i])
                    review_review.append((review[l].text).replace('\n',''))
                    review_star.append(star[l].text)
                    review_name.append(name[l].text)
                    review_date.append(date[l].text)
                    kkuk_lst.append(kkuk[l].text)
                    # 얼마나 했는지 확인
                    if l % 100 == 0:
                        print(l)
                except:
                    product_name.append(None)
                    review_star.append(None)
                    review_name.append(None)
                    review_date.append(None)
                    review_review.append(None)
                    kkuk_lst.append(None)
                    continue

            print(i,"번 완료")
            driver.quit()
        except: # 리뷰가 없어서 안 눌릴때 driver 닫고 계속
            driver.quit()
            continue
    except:
        driver.quit()
        continue
    
    if i % 100 == 0:
        data = pd.DataFrame([ x for x in zip(product_name, review_date,review_name,review_star,review_review,kkuk_lst)],columns = ['상품명','날짜','ID','평점','리뷰','꾹'])    
        data.to_csv(f"11번가_1점리뷰_0827_{i}.csv", encoding = "utf-8-sig")
        
```
