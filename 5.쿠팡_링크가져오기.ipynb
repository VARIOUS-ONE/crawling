```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
import random
from selenium.webdriver.common.keys import Keys
```

### 처음 60개 링크 가져오기(60개

```python
url = "https://www.coupang.com/np/campaigns/82?listSize=60&brand=&offerCondition=&filterType=&isPriceRange=false&minPrice=&maxPrice=&page=1&channel=user&fromComponent=N&selectedPlpKeepFilter=&sorter=bestAsc&filter=&rating=0"

driver = webdriver.Chrome("chromedriver.exe")
driver.get(url)
driver.implicitly_wait(5)

product_name = []
product_star = []
product_link = []
review_count = []

# 맨 아래로 내려가기
driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")

#XPath로 제품 이름과 별점 크롤
for i in range(60): #60
    name1 = driver.find_element(By.XPATH, f"/html/body/div[3]/section/form/div/div/div[1]/div[4]/ul/li[{i+1}]/a/dl/dd/div[2]").text
    star1 = driver.find_element(By.XPATH, f"/html/body/div[3]/section/form/div/div/div[1]/div[4]/ul/li[{i+1}]/a/dl/dd/div[4]/div/span[1]/em").text
    
    product_name.append(name1)
    product_star.append(star1)
    

link1 = driver.find_elements(By.CLASS_NAME, "baby-product-link")
count1 = driver.find_elements(By.CLASS_NAME, "rating-total-count")

for i in range(60): # 60
    product_link.append(link1[i].get_attribute('href'))
    review_count.append(count1[i].text)
driver.quit()
```

```python
import pandas as pd
```

```python
data = pd.DataFrame([ x for x in zip(product_name, product_link, product_star , review_count)],columns = ['상품명','링크','평균별점','리뷰수'])    

data
```

```python
data.info()
```

### 리뷰수 평균별점 숫자로 바꾸기

```python
for i in range(len(data)):
    data['리뷰수'][i] = data['리뷰수'][i][1:-1]
```

```python
data['평균별점'] = data['평균별점'].astype('float')
data['리뷰수'] = data['리뷰수'].astype('int')
```

```python
data.head()
```

```python
data.to_csv("쿠팡링크.csv", encoding = "utf-8-sig")
```
