```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import random

import pandas as pd
```

# 1점대 크롤링

```python
review_star = []
review_name = []
review_date = []
review_review = []
kkuk_lst = []

url = "https://www.11st.co.kr/products/1929823447?inpu=&trTypeCd=22&trCtgrNo=895019"
# 리뷰 5000개짜리

driver = webdriver.Chrome("chromedriver.exe")
driver.get(url)
driver.implicitly_wait(3)

# iframe 때문에 에러 뜸
# frame은 총 4개 있고 그 중 ifrmReview를 찾아내서
# ifrmReview인 곳에 switch 하기
iframe = driver.find_element(By.ID, "ifrmReview") 
driver.switch_to.frame(iframe)

# 평점 고르는 버튼 누르기
driver.find_element(By.XPATH, "/html/body/div[2]/ul/li[1]/div[1]/div[1]/button").send_keys(Keys.ENTER)
time.sleep(random.uniform(0,2))

# 평점 1점 누르기
driver.find_element(By.XPATH, "/html/body/div[2]/ul/li[1]/div[1]/div[2]/dl/dd[6]/label/span").click()
time.sleep(random.uniform(0,2))
    
    #  끝날 때까지 더 보기 누르기
p = 0
while True:
    try:
        driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/button").send_keys(Keys.ENTER)
        # driver.implicitly_wait(3)
        p = p + 1
        # 리뷰 1000개당 15초 쉬기
        if p % 50 == 0:
            driver.execute_script("window.scrollBy(0,250);")
            time.sleep(15)
            print(p, "번째 스크롤 끝")
    # 더 누를 더보기 버튼이 없으면 몇번 눌렀는지 나오고 while문 빠져나오기
    except:
        print(f"총 {p}번 더보기 버튼 누름")
        break

# try:
star = driver.find_elements(By.CSS_SELECTOR,'p.grade > span > em')
name = driver.find_elements(By.CLASS_NAME, "name")
date = driver.find_elements(By.CLASS_NAME, "date")
review = driver.find_elements(By.CLASS_NAME, "cont_text_wrap")
kkuk = driver.find_elements(By.CLASS_NAME,'kkukCount')

# 리뷰 내용이 없을 수 있으므로
# try except 문으로 처리하기
for i in range(len(star)):
    try:
        review_review.append(review[i].text)
        review_star.append(star[i].text)
        review_name.append(name[i].text)
        review_date.append(date[i].text)
        kkuk_lst.append(kkuk[i].text)
        # 얼마나 했는지 확인
        if i % 100 == 0:
            print(i)
    except:
        review_star.append(None)
        review_name.append(None)
        review_date.append(None)
        review_review.append(None)
        kkuk_lst.append(None)
        continue


driver.quit()  
```

```python
data = pd.DataFrame([ x for x in zip(review_date,review_name,review_star,review_review,kkuk_lst)],columns = ['날짜','ID','평점','리뷰','꾹'])    
data
```
